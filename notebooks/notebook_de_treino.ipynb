```python
df.describe(include="all") # A diferença desse .describe() é o parâmetro include="all", esse parâmetro traz todas as informações, enquanto o anterior traz apenas as informações numéricas. Consequentemente ele traz muitos dados como NaN (nulos), porém não são nulos de verdade.
```

```python
df.describe(include="object") # Com esse describ, deferente do "all", conseguimos trazer apenas as que são colunas com textos.
```

```python
df.describe(percentiles=[.01, .1, .25, .5, .75, .9, .95, .99]) # Com esse describe, coneguimos modificar os percentis. O padrão é apenas 25%, 50% e 75%. Com esse colocamos mais percentis.

# Isso é útil para detectar outliers extremos. O percentil 99 te diz onde estão os 1% maiores valores — muito usado em análise de fraude, por exemplo.
```
